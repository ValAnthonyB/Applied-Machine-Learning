```python
import matplotlib.pyplot as plt
import numpy as np
import os
import tensorflow as tf

from tensorflow.keras.preprocessing import image_dataset_from_directory
from tensorflow.keras.layers.experimental.preprocessing import RandomFlip, RandomRotation, RandomContrast, RandomZoom
```

# Load the data

```python
BATCH_SIZE = 10
IMG_SIZE = (1024, 1024)

train_dataset = image_dataset_from_directory("C:/Users/tonba/OneDrive/Desktop/Machine Learning Projects/datasets/COVID-19 Dataset/train",
                                             shuffle=True,
                                             batch_size=BATCH_SIZE,
                                             image_size=IMG_SIZE,
                                             seed=42)

test_dataset = image_dataset_from_directory("C:/Users/tonba/OneDrive/Desktop/Machine Learning Projects/datasets/COVID-19 Dataset/test",
                                             shuffle=True,
                                             batch_size=BATCH_SIZE,
                                             image_size=IMG_SIZE,
                                             seed=42)

# Class names
class_names = train_dataset.class_names
print(class_names)
```

##### View the images

# Data augmentation

```python
# Use horizontal flip and random rotation techniques for data augmentation
def data_augmenter():
    '''
    Create a Sequential model composed of 2 layers
    Returns:
        tf.keras.Sequential
    '''
    data_augmentation = tf.keras.Sequential()
    #data_augmentation.add(RandomFlip('horizontal')) do not include this assuming that we do not want to move the heart's position
    data_augmentation.add(RandomRotation(0.2))
    data_augmentation.add(RandomContrast(.3))
    data_augmentation.add(RandomZoom(0.2))  
    
    return data_augmentation

# This will be used in parallel with our final model
data_augmentation = data_augmenter() 
```

# Apply prefetch to the training and validation set

```python
AUTOTUNE = tf.data.experimental.AUTOTUNE

train_dataset = train_dataset.prefetch(buffer_size=AUTOTUNE)
test_dataset = test_dataset.prefetch(buffer_size=AUTOTUNE)
```

# Model

```python
def COVID_recognition_model():
    
    # Use the ResNet50 model in Keras as the base model
    base = tf.keras.applications.resnet50.ResNet50(include_top=False, 
                                                  weights='imagenet', 
                                                  input_shape=(1024, 1024)+(3,))
    
    # Freeze early layers until layer 40
    for base_layer in base.layers[:155]:
        base_layer.trainable = False
    
    #__________________________________________________________________________
    # Input layer
    input_ = tf.keras.layers.Input(shape=(1024, 1024)+(3,))
    
    # Layer for rescaling
    X = tf.keras.layers.Rescaling(1./255)(input_)
    
    # Data augmentation
    X = data_augmentation(X) 
    
    # Use the ResNet50 architecture as the model
    X = base(X, training=False)
    
    # Average pooling
    X = tf.keras.layers.AveragePooling2D(pool_size=(4, 4))(X)
    
    # Flatten
    X = tf.keras.layers.Flatten()(X)
    
    # Add dropout
    X = tf.keras.layers.Dropout(0.15)(X)
    
    # Prediction layer
    prediction_layer = tf.keras.layers.Dense(1, activation='sigmoid')(X)
    
    # Defines the final model
    model = tf.keras.Model(inputs=input_, outputs=prediction_layer)
    
    return model


model = COVID_recognition_model()
model.summary()
```

# Compile the model

```python
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

# Train the network

```python
history = model.fit(train_dataset, 
                    epochs=5, 
                    validation_data=test_dataset,
                    batch_size=10
                   )
```

# Model evaluation

```python
model.evaluate(test_dataset)
```

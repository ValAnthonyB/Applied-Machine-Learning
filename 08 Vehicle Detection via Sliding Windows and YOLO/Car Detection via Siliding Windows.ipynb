```python
import matplotlib.pyplot as plt
import numpy as np
import os

import tensorflow as tf
```

# Load the dataset

```python
from tensorflow.keras.preprocessing import image_dataset_from_directory

batch_size = 128
image_size = (64, 64)
directory = "C:/Users/tonba/OneDrive/Desktop/Machine Learning Projects/datasets/car dataset"

train_ds, val_ds = tf.keras.preprocessing.image_dataset_from_directory(
                                                                        directory,
                                                                        validation_split=0.3,
                                                                        subset="both",
                                                                        seed=1,
                                                                        shuffle=True,
                                                                        image_size=image_size,
                                                                        batch_size=batch_size,
                                                                        )

# Class names
class_names = train_ds.class_names
print(class_names)
```

# Display the images

```python
plt.figure(figsize=(10, 10))
for images, labels in train_ds.take(2):
    for i in range(9):
        ax = plt.subplot(3, 3, i + 1)
        plt.imshow(images[i].numpy().astype("uint8"))
        plt.title(class_names[labels[i]])
        plt.axis("off")
```

# Apply horizontal flip and random flip data augmentation

```python
from tensorflow.keras.layers.experimental.preprocessing import RandomFlip, RandomRotation

def data_augmenter():
    '''
    Create a Sequential model composed of 2 layers
    Returns:
        tf.keras.Sequential
    '''
    data_augmentation = tf.keras.Sequential()
    data_augmentation.add(RandomFlip('horizontal'))
    data_augmentation.add(RandomRotation(0.2))
    
    return data_augmentation

# This will be used in parallel with our final model
data_augmentation = data_augmenter() 
```

# MobileNetV2 model in Keras with ImageNet weights as the base model

```python
def final_model():
    # Use the MobileNetV2 model in Keras as the base model
    base = tf.keras.applications.MobileNetV2(include_top=False, 
                                                  weights='imagenet', 
                                                  input_shape=image_size+(3,))
    
    #__________________________________________________________________________
    # Input layer
    input_ = tf.keras.layers.Input(shape=image_size+(3,))
    
    # Layer for rescaling
    X = tf.keras.layers.Rescaling(1./255)(input_)
    
    # Data augmentation
    X = data_augmentation(X) 
    
    # Use the MobileNetV2 architecture as the model
    X = base(X)
    
    # Flatten
    X = tf.keras.layers.Flatten()(X)
    
    # Add dropout
    X = tf.keras.layers.Dropout(0.3)(X)
    
    # Prediction layer
    prediction_layer = tf.keras.layers.Dense(1, activation='sigmoid')(X)
    
    # Defines the final model
    model = tf.keras.Model(inputs=input_, outputs=prediction_layer)
    
    return model

model = final_model()
model.summary()
```

# Compile the model

```python
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

# Train the model

```python
history = model.fit(train_ds, 
                    epochs=5, 
                    validation_data=val_ds)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import os
import tensorflow as tf
import tensorflow.keras.layers as tfl

from tensorflow.keras.preprocessing import image_dataset_from_directory
from tensorflow.keras.layers.experimental.preprocessing import RandomFlip, RandomRotation
```

# Problem statement

We want to make an Alpaca image recognition model that uses the MobileNetV2 weights pre-trained on ImageNet.

### Load the dataset from directory

```python
BATCH_SIZE = 32
IMG_SIZE = (160, 160)
directory = "C:/Users/tonba/OneDrive/Desktop/Machine Learning Projects/datasets/alpaca dataset/"

train_dataset = image_dataset_from_directory(directory,
                                             shuffle=True,
                                             batch_size=BATCH_SIZE,
                                             image_size=IMG_SIZE,
                                             validation_split=0.2,
                                             subset='training',
                                             seed=42)

validation_dataset = image_dataset_from_directory(directory,
                                             shuffle=True,
                                             batch_size=BATCH_SIZE,
                                             image_size=IMG_SIZE,
                                             validation_split=0.2,
                                             subset='validation',
                                             seed=42)
```

##### View some of the images in the training set

Since we set `batch_size=32` then the training set will have 8 batches. We use `train_dataset.take(1)` to get the first batched training data.

```python
class_names = train_dataset.class_names

plt.figure(figsize=(10, 10))
for images, labels in train_dataset.take(8):
    for i in range(9):
        ax = plt.subplot(3, 3, i + 1)
        plt.imshow(images[i].numpy().astype("uint8"))
        plt.title(class_names[labels[i]])
        plt.axis("off")
```

##### Applies prefetch to the training dataset which prevents memory bottleneck (not enough memory to store training data)

```python
AUTOTUNE = tf.data.experimental.AUTOTUNE
train_dataset = train_dataset.prefetch(buffer_size=AUTOTUNE)
```

### Data augmentation

```python
# Use horizontal flip and random rotation techniques for data augmentation
def data_augmenter():
    '''
    Create a Sequential model composed of 2 layers
    Returns:
        tf.keras.Sequential
    '''
    data_augmentation = tf.keras.Sequential()
    data_augmentation.add(RandomFlip('horizontal'))
    data_augmentation.add(RandomRotation(0.2))
    
    return data_augmentation

# This will be used in parallel with our final model
data_augmentation = data_augmenter() 
```

### Construct the modified model

1. Import the MobileNetV2 model with the ImageNet weights and use it as the base model
2. Set include_top=False because we are not performing the ImageNet problem
3. Set the prediction (last layer) as a single dense layer and apply the Sigmoid activation function
4. Select the `fine_tune_at` layer and freeze the preceding layers
5. Train the remaining layers

```python
def alpaca_model(image_shape=IMG_SIZE, data_augmentation=data_augmenter()):
    ''' Define a tf.keras model for binary classification out of the MobileNetV2 model
    Arguments:
        image_shape -- Image width and height
        data_augmentation -- data augmentation function
    Returns:
    Returns:
        tf.keras.model
    '''
    
    # We set include_top=False because we are not doing the ImageNet problem.
    # We modify the last layers to our own needs.
    input_shape = image_shape + (3,)
    preprocess_input = tf.keras.applications.mobilenet_v2.preprocess_input 
    base_model = tf.keras.applications.MobileNetV2(input_shape=input_shape,
                                                           include_top=False, 
                                                           weights="imagenet") # From imageNet
    
    # Fine-tune from this layer onwards
    fine_tune_at = 120

    # Freeze all the layers before the `fine_tune_at` layer
    for layer in base_model.layers[:fine_tune_at]:
        layer.trainable = False
    
    # Layers start here
    ####_______________________________________________________________________________________________________
    
    # Create the input layer (Same as the imageNetv2 input size)
    inputs = tf.keras.Input(shape=input_shape) 
    
    # apply data augmentation to the inputs
    X = data_augmentation(inputs) 
    
    # data preprocessing using the same weights the model was trained on
    X = preprocess_input(X) 
    
    # set training to False to avoid keeping track of statistics in the batch norm layer
    X = base_model(X, training=False) 
    
    # Add the new Binary classification layers
    # Use global avg pooling to summarize the info in each channel
    X = tfl.GlobalAveragePooling2D()(X)
    
    # include dropout with probability of 0.2 to avoid overfitting
    X = tfl.Dropout(0.15)(X)
        
    # Use a prediction layer with one neuron (as a binary classifier only needs one)
    prediction_layer = tfl.Dense(1, activation='sigmoid')
    outputs = prediction_layer(X)
    
    # Final model
    model = tf.keras.Model(inputs, outputs)
    
    return model
```

### Compile and train the model

```python
model = alpaca_model(IMG_SIZE, data_augmentation)
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
              loss=tf.keras.losses.BinaryCrossentropy(),
              metrics=['accuracy'])

initial_epochs = 10
history = model.fit(train_dataset, validation_data=validation_dataset, epochs=20)
```

### Predict the test set

Let's take the first batch of the validation set and make predictions.

```python
predictions = model.predict(validation_dataset)
predictions = tf.where(predictions < 0.5,0, 1)
#print('Predictions:\n', predictions.numpy())
```

### Model evaluation

```python
model.evaluate(validation_dataset)
```
